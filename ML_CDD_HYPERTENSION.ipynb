## *search for the TARGETING PROTEIN*

Target search for the ***hypertension*** ***Angiotensin II***

```python
target = new_client.target
target_query = target.search('Angiotensin II')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
targets
```

selecting the **bioactivity data** of the ***AGT*** in ***hypertension*** ***(1 ENTRY)***

```python
selected_target = targets.target_chembl_id[15]
selected_target
```

```python
# Searching for the specific chembl id in the above results
chembl_id = targets.loc[targets['target_chembl_id'] == 'CHEMBL227']

# Print the specific chembl id
print(chembl_id)
```

Here, we will retrieve only bioactivity data for AGT  RECEPTOR (CHEMBL3596085) that are reported as IC 50  values in nM (nanomolar) unit.

filtering of the IC 50 values ⏳

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

# Handling missing **data**

If any compounds has missing value for the standard_value column then drop it

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

Data pre-processing of the bioactivity data


Labeling compounds as either being active, inactive or intermediate

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
df2.molecule_chembl_id
```

Iterate the molecule_chembl_id to a list

Iterate the molecule_chembl_id to a list

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

### **Iterate canonical_smiles to a list** 

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

**Iterate standard_value to a list**

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

## **Combine the 4 lists into a dataframe**

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
